```python
import pandas as pd 


dataFrameDados = pd.read_excel(r"C:\\Users\\thais\\Documents\\Udemy - Curso de Python com RPA e Criação de Projetos Reais\\Python e Pandas - Análise e Tratamento de Dados\\Planilhas\\Deletar_Linhas_Colunas.xlsx")

display(dataFrameDados)

deletandoLinhasemBranco= dataFrameDados.dropna()

#DROP = Deleta uma ou mais colunas
deletarDuasColunas= deletandoLinhasemBranco.drop(columns=["Produto", "Data Venda"])

#AXIS = eixo(1 - Coluna - 0 - Linha)
excluirLinha3 = deletarDuasColunas.drop(2,axis=0)

excluirLinha01 = deletarDuasColunas.drop([0,1])


display(excluirLinha3)
display(excluirLinha01)
```
