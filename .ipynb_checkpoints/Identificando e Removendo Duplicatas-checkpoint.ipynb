```python
import pandas as pd

baseVendas_DataFrame = pd.read_excel(r"C:\Users\thais\Documents\Udemy - Curso de Python com RPA e Criação de Projetos Reais\Python e Pandas - Análise e Tratamento de Dados\Planilhas\Base_Vendas.xlsx")

display(baseVendas_DataFrame)

#NUNIQUE =  Resume valores deixando somente valores unicos
resumoValoresUnicos = baseVendas_DataFrame.nunique()

display(resumoValoresUnicos)

#DUPLICATE = Identificando valores duplicados

# SUBSET = Identifica a coluna qe queremos verificar a duplicidade

# KEEP = Controla como considerar o valor duplicado(primeiro, ultimo e false considera todos)


confereDuplicidades = baseVendas_DataFrame.duplicated(subset="Vendedor", keep="first")

display(confereDuplicidades)
```
