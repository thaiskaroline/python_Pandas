```python
import pandas as opcoesPanda

notasAlunos_DataFrame = opcoesPanda.DataFrame({
    "Nome":["Luiza", "Magali", "Cascão"],
    "Nota 1": [9,8,8.9],
    "Nota 2": [7.5,8.2,9.1],
    "Nota 3": [5,2,5],
    "Nota 4": [6.5,2.2,3.1]
})

display(notasAlunos_DataFrame)


#Adicionar nova coluna

notasAlunos_DataFrame["Média"] = (notasAlunos_DataFrame["Nota 1"] + notasAlunos_DataFrame["Nota 2"] + notasAlunos_DataFrame["Nota 3"] + notasAlunos_DataFrame["Nota 4"]) / 4

novaColunaFaltas = [5,6,2]


#Criando coluna com valor padrão

notasAlunos_DataFrame["Faltas"] = novaColunaFaltas


#Alterando dados do dataFrame

#LOC = LOCALIZAR

notasAlunos_DataFrame.loc[1,"Nota 2"] = 10


display(notasAlunos_DataFrame)
```
